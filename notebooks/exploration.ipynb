```python
import hopsworks

project = hopsworks.login()
fs = project.get_feature_store()

air_quality_fg = fs.get_feature_group(name="air_quality_daily", version=1)

feature_view = fs.get_or_create_feature_view(
    name="air_quality_daily_fv",
    version=1,
    feature_group=air_quality_fg
)

df = feature_view.get_batch_data()
print("Rows and Columns:", df.shape)
df.head()
```
